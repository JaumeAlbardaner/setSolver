```python
from ultralytics import YOLO
```

```python
# Load the model.
model = YOLO('yolov8n.pt')
 
# Training.
results = model.train(
   data='setCards.yaml',
   imgsz=640,
   epochs=100,
   batch=40,
   name='yolov8n_setSolver')
```

After training, now we run the model on the validation set to observe performance

```python
from ultralytics import YOLO

model = YOLO('runs/detect/yolov8n_setSolver2/weights/best.pt')
model.predict('Validation_Dataset', save=True, imgsz=640, conf=0.2)
```

Convert model to tflite for usage in Android

```python
from ultralytics import YOLO

model = YOLO('runs/detect/yolov8n_setSolver2/weights/best.pt')

model.export(format="tflite")
```
